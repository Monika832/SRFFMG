```python
import os
import random
import shutil

def split_images(source_folder, folder1, folder2, folder3, ratio):
    # Create the destination folders if they don't exist
    os.makedirs(folder1, exist_ok=True)
    os.makedirs(folder2, exist_ok=True)
    os.makedirs(folder3, exist_ok=True)
    
    # Get a list of all image files in the source folder
    image_files = [file for file in os.listdir(source_folder) if file.endswith(('.jpg', '.jpeg', '.png'))]
    
    # Shuffle the image files
    random.shuffle(image_files)
    
    total_images = len(image_files)
    num_folder1 = int(total_images * ratio[0] / 100)
    num_folder2 = int(total_images * ratio[1] / 100)
    
    # Copy images to folder 1
    for image in image_files[:num_folder1]:
        source_path = os.path.join(source_folder, image)
        destination_path = os.path.join(folder1, image)
        shutil.copyfile(source_path, destination_path)
        print(f"Copied to folder 1: {image}")
    
    # Copy images to folder 2
    for image in image_files[num_folder1:num_folder1+num_folder2]:
        source_path = os.path.join(source_folder, image)
        destination_path = os.path.join(folder2, image)
        shutil.copyfile(source_path, destination_path)
        print(f"Copied to folder 2: {image}")
    
    # Copy images to folder 3
    for image in image_files[num_folder1+num_folder2:]:
        source_path = os.path.join(source_folder, image)
        destination_path = os.path.join(folder3, image)
        shutil.copyfile(source_path, destination_path)
        print(f"Copied to folder 3: {image}")
    
    print("Image splitting completed.")

# Example usage
source_folder = "celeba_hq_resized"
folder1 = "Dataset/train_Set"
folder2 = "Dataset/test_Set"
folder3 = "Dataset/val_Set"
ratio = [70, 20, 10]

split_images(source_folder, folder1, folder2, folder3, ratio)
```

```python
def resize_images(source_folder, destination_folder, target_size):
    # Create the destination folder if it doesn't exist
    os.makedirs(destination_folder, exist_ok=True)

    # Get a list of all image files in the source folder
    image_files = [file for file in os.listdir(source_folder) if file.endswith(('.jpg', '.jpeg', '.png'))]

    for image_file in image_files:
        # Read the image
        image_path = os.path.join(source_folder, image_file)
        image = cv2.imread(image_path)

        # Resize the image
        resized_image = cv2.resize(image, target_size)

        # Save the resized image to the destination folder
        destination_path = os.path.join(destination_folder, image_file)
        cv2.imwrite(destination_path, resized_image)
        print(f"Resized and saved: {image_file}")

    print("Image resizing and saving completed.")
```

```python
def apply_gaussian_blur(source_directory, destination_directory, k):
    # Create the destination directory if it doesn't exist
    os.makedirs(destination_directory, exist_ok=True)
    
    # Get a list of all image files in the source directory
    image_files = [file for file in os.listdir(source_directory) if file.endswith(('.jpg', '.jpeg', '.png'))]
    
    # Apply Gaussian blur to each image and save in the destination directory
    for image_file in image_files:
        source_path = os.path.join(source_directory, image_file)
        destination_path = os.path.join(destination_directory, image_file)
        
        # Read the image
        image = cv2.imread(source_path)
        
        # Apply Gaussian blur with kernel size k
        blurred_image = cv2.GaussianBlur(image, (k, k), 0)
        
        # Save the blurred image
        cv2.imwrite(destination_path, blurred_image)
        
        print(f"Blurred image saved: {destination_path}")

    print("Gaussian blur applied to all images.")
```

```python
source_directory = "Dataset/test_Set/HR"
destination_directory = "test_Outputs/3"
k = 3

apply_gaussian_blur(source_directory, destination_directory, k)
```
